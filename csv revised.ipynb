```python
import pandas as pd
import re

df = pd.read_csv('data_science_jobs_indeed_usa.csv')
```

```python
df
```

```python
grp = pd.pivot_table(df,
                    index='Title',
                    aggfunc='count')

grp
```

```python
grp1 = grp.iloc[:, [0]] # [:, 0]을 하면 type이 series로 나옴
grp1 = grp1.rename(columns={'Company': 'count'})
```

```python
print(grp1.sort_values(by='count', ascending=False))
```

```python
# 조건에 따라 새로운 열 추가
def title_cat(x):
    if re.search(r'\bData Engineer\b', x):
        return 'Data Engineer'
    if re.search(r'\bBusiness Analyst\b', x):
        return 'Business Analyst'
    if re.search(r'\bData Scientist\b', x):
        return 'Data Scientist'
    if re.search(r'\bBusiness Intelligence Developer|BI Developer\b', x):
        return 'BI Developer'
    if re.search(r'\bData Analyst\b', x):
        return 'Data Analyst'
    if re.search(r'\bManager\b', x):
        return 'Manager'
    if re.search(r'\bQA\b', x):
        return 'QA'
    if re.search(r'\bMachine Learning Engineer\b', x):
        return 'Machine Learning Engineer'
    else:
        return 'others'

df['title_cat'] = df['Title'].apply(lambda x : title_cat(x))

df
```
